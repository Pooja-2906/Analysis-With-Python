# Data Visualization
# Histogram

```python
count=df['Gender'].value_counts()
count
```

```python
plt.hist(df['Gender']) #---- Mam Suggestion - Total Sleep Hours VS Sleep quality
plt.yticks(count)
plt.yticks(rotation=10)
plt.title('Gender',c="Red")
plt.show()
```

# Bar Plot

```python
# Bar Plot -----> Compare Categorical vs numerical values
# Gender VS Sleep Quality
```

```python
x=df['Gender']
y=df['Sleep Quality']
plt.bar(x,y,color="Violet")
plt.title("Sleep Quality VS Gender")
plt.yticks(y)
plt.show()
```
